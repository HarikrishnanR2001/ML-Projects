```python
# function to import data
import os
import tarfile
import urllib

DOWNLOAD_ROOT="https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH=os.path.join("datasets","housing")
HOUSING_URL=DOWNLOAD_ROOT+"datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL,housing_path=HOUSING_PATH):
    os.makedirs(housing_path,exist_ok=True)
    tgz_path=os.path.join(housing_path,"housing.tgz")
    urllib.request.urlretrieve(housing_url,tgz_path)
    housing_tgz=tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
fetch_housing_data()
```

```python
import numpy as np

def datashuffle(data,ratio):
    shuffled_indices=np.random.permutation(len(data))
    test_no=int(len(data)*ratio)
    test_data_indices=shuffled_indices[:test_no]
    train_data_indices=shuffled_indices[test_no:]
    return data.iloc[test_data_indices],data.iloc[train_data_indices]

(test_set,train_set)=datashuffle(data,0.2)
```

```python
train_val=strat_train_set.drop("median_house_value",axis=1)
target_val=strat_train_set["median_house_value"].copy()
train_val.head()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
imputer.fit(train_val)
x=imputer.transform(train_val)
```

```python
train_val=pd.DataFrame(x,columns=train_val.columns,index=train_val.index)
train_val.head()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

# column index positions (after imputer step  DataFrame  NumPy array)
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        # nothing to learn here
        return self
    
    def transform(self, X):
        # engineered features
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            # add new features as extra columns
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]


# usage
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(train_val.to_numpy())
```

```python
housing_extra_attribs
```

```python
# Creating a pipeline of transformers using sklearn

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline=Pipeline([
 ("imputer",SimpleImputer(strategy="median")),
 ("attribs_adder",CombinedAttributesAdder()),
 ("std_scaler",StandardScaler()),
 ])
train_val.head()
cleaned_train_val=num_pipeline.fit_transform(train_val.to_numpy())

 
```

```python
 # Adding back column names

column_names =["longitude","latitude","housing_median_age","total_rooms","total_bedrooms","population","households","median_income","rooms_per_house","bedrooms_per_room","population_per_household"]
column_names=column_names+["rooms_per_household", "population_per_household", "bedrooms_per_room"]
cleaned_data=pd.DataFrame(cleaned_train_val,columns=column_names)
# cleaned_data.head()
train_val.head()
```

```python
# fitting Linear Regression model
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(cleaned_data,target_val)
```

```python
# Taking values from the trainning set
some_data=cleaned_data.iloc[:5]
some_labels=target_val.iloc[:5]
print(lin_reg.predict(some_data))
print(list(some_labels))
```

```python
# calculating performance of the model
from sklearn.metrics import mean_squared_error,r2_score
predictions=lin_reg.predict(cleaned_data)
lin_mse=mean_squared_error(target_val,predictions)
lin_rmse=np.sqrt(lin_mse)
r2=r2_score(target_val,predictions)
print(f"R square value:{r2}")
print(f"RMSE value:{lin_rmse}")
```

```python
#fitting decision tree:
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(cleaned_data,target_val)
```

```python
predictions=tree.predict(cleaned_data)
tree_mse=mean_squared_error(target_val,predictions)
tree_rmse=np.sqrt(tree_mse)
print(f"RMSE:{tree_rmse}")
```

```python
#using cross validation for checking desicion tree RMSE 
from sklearn.model_selection import cross_val_score
scores=cross_val_score(tree,cleaned_data,target_val,scoring="neg_mean_squared_error",cv=10)
tree_rmse_scores=np.sqrt(-scores)
print(tree_rmse_scores)
print("Perfoemance metrics for decision tree")
print(f"MEAN RMSE: {tree_rmse_scores.mean()},SD OF RMSE: {tree_rmse_scores.std()} ")
print()
```

```python
scores=cross_val_score(lin_reg,cleaned_data,target_val,scoring="neg_mean_squared_error",cv=10)
lin_rmse_scores=np.sqrt(-scores)
print(lin_rmse_scores)
print(f"MEAN RMSE: {lin_rmse_scores.mean()},SD OF RMSE: {lin_rmse_scores.std()} ")
print()
```
